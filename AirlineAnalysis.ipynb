```python
data['arrival_time'].value_counts(ascending=True)
```

```python
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.bar(data['departure_time'].value_counts().index, data['departure_time'].value_counts().values, color=['violet','maroon'])
plt.title("Departure Time")
plt.xlabel("Departure Timings")
plt.ylabel("Departure Flights")


plt.subplot(1,2,2)
plt.bar(data['arrival_time'].value_counts().index, data['arrival_time'].value_counts().values, color=['Green','Orange'])
plt.title("Arrival Time")
plt.xlabel("Arrival Timings")
plt.ylabel("Arrival Flights")
plt.tight_layout()
plt.show()
```
